## 1.1. A estrutura dos dados

Os dados tem origem no livro "Machine Learning with R" Brett Lantz, a publicação fornece uma introdução ao aprendizado de máquina. Os dados tratam de registros sobre beneficiários de um seguro de saúde nos EUA e vamos utilizá-los para fazer predições e avaliar os fatores que afetam o preço deste produto através da biblioteca Scikit-learn.

#### Os dados seguem a seguinte estrutura:
|Coluna|Descrição|
|-------|---------|
|age|Idade do beneficiário|
|sex|Genêro|
|bmi|O IMC (Índice de Massa Corporal) do beneficiário|

|region|A zona residencial nos EUA do beneficiário|
|charges|Os custos individuais do seguro|

# 4. Regressão Linear

## 4.1. Pré-processamento dos dados

### Transformação logarítimica

Como observado durante a análise exploratória, a distribuição de 'charges' tem uma assimétria. Vou aplicar uma transformação antes de prosseguir. 

```python
df_transformado = df.copy()
df_transformado['charges'] = np.log(df_transformado['charges'])
df_transformado.head()
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.add_axes([0, 0, 1, 1])

sns.histplot(data = df_transformado, x = 'charges', kde = True)
ax.set_title('Distribuição dos Custos', loc = 'left', fontsize = 15, pad = 10)
ax.set_xlabel('log(Custos em US$)', fontsize = 10)
ax.set_ylabel('Contagem', fontsize = 10)
plt.show()
```

#### Observações

Agora é possível ver uma distribuição mais próxima à normal. 

### Definindo os conjuntos $X$ e $y$

```python
# Features
X = df_transformado.drop(columns = 'charges', axis = 1)
# Target
y = df_transformado['charges']
```

### Tratando as variáveis categóricas com OneHotEncoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

cat_features = ['sex', 'smoker', 'region']
column_transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(), cat_features)
], remainder = 'passthrough')

X_encoded = column_transformer.fit_transform(X)
column_names = column_transformer.get_feature_names_out()
X_encoded_df = pd.DataFrame(X_encoded, columns=column_names)
```

### Matriz de Correlação

```python
data = pd.concat([X_encoded_df, y], axis = 1)
data.head()
```

```python
correlacao = data.corr()
correlacao
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.add_axes([0, 0, 1, 1])

sns.heatmap(correlacao, annot = True, fmt = '.2f', square = True, cmap = my_palette, linecolor = 'white', linewidth = 0.5)
ax.set_title('Matriz de Correlação', loc = 'left', fontsize = 16, pad = 12)
plt.show()
```

### Definindo conjuntos de treino e teste

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded_df, y, test_size = 0.20, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 4.3. LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo_lr = LinearRegression()
```

```python
modelo_lr.fit(X_train, y_train)
```

```python
# Previsão
y_pred = modelo_lr.predict(X_test)
y_pred[0]
```

```python
np.exp(y_pred[0])
```

### Coeficientes da Regressão Linear

```python
index = ['Intercepto', 'Sex_Female', 'Sex_Male', 'Smoker_No', 'Smoker_Yes', 'Region_Northeast', 
         'Region_Northwest', 'Region_Southeast', 'Region_Southwest', 'Age', 'BMI', 'Children']
```

```python
pd.DataFrame(data = np.append(modelo_lr.intercept_, modelo_lr.coef_), index = index, columns = ['Coeficientes'])
```

```python
print(f"O intercepto é de US${np.exp(7.713988).round(2)}")
```

#### Sobre os coeficientes

Como apliquei a transformação logarítimica a interpretação dos coeficentes muda um pouco. Vamos observar o exemplo de Age.

$log(Charges) = {\beta}_0 + {\beta}_{Age} \times Age$

$log(Charges) = 7.713988 + 0.33997 \times Age$

$Charges = e^{log(Charges)}$

$Charges = e^{7.713988+0.033997 \times Age}$

$Charges = e^{7.713988} \times e^{+0.033997 \times Age}$

Essa expressão representa o modelo de regressão onde a variável $y$ ou 'charges' é modelada como uma função exponencial de 'age', com um intercepto em log. Dessa forma para cada unidade acrescida em 'age' os custos do seguro, assumindo todos os outros fatores constantes, terá um aumento de 3,4% ($e^{0.033997} \approx 1.034$).

### Métricas

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {(r2 * 100).round(2)}% ")
```

```python
y_pred_train = modelo_lr.predict(X_train)
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.add_axes([0, 0, 1, 1])

sns.regplot(x = y_train, y = y_pred_train)
ax.set_title('Real x Previsto', loc = 'left', fontsize = 15, pad = 10)
ax.set_xlabel('log(Custos em US$)', fontsize = 8)
plt.show()
```

Esse dois fatores são os de maior efeito nos custos.
